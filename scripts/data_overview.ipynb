```python
import pandas as pd
import matplotlib.pyplot as plt

path = "../data/1-raw/Daily_Water_Intake.csv"
df = pd.read_csv(path)
```

```python
# Plot Age
plt.figure()
plt.hist(df["Age"].dropna())
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Distribution of Age")
plt.show()
```

```python
# Plot Weight
plt.figure()
plt.hist(df["Weight"].dropna())
plt.xlabel("Weight")
plt.ylabel("Count")
plt.title("Distribution of Weight")
plt.show()
```

```python
# Plot Gender
plt.figure()
plt.hist(df["Gender"].dropna())
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Distribution of Gender")
plt.show()
```

```python
# Count occurrences
counts = (
    df.groupby(["Weather", "Hydration Level"])
      .size()
      .unstack(fill_value=0)
)

# Plot stacked bars
counts.plot(
    kind="bar",
    stacked=True,
    color=["blue", "red"]
)

plt.xlabel("Weather")
plt.ylabel("Count")
plt.title("Weather vs Hydration Level")
plt.legend(title="Hydration Level")
plt.show()
```

```python
# Plot Weight
plt.figure()
plt.hist(df["Physical Activity Level"].dropna())
plt.xlabel("Physical Activity Level")
plt.ylabel("Count")
plt.title("Distribution of Physical Activity Level")
plt.show()
```

```python
# Plot Hydration Level
plt.figure()
plt.hist(df["Hydration Level"].dropna())
plt.xlabel("Hydration Level")
plt.ylabel("Count")
plt.title("Distribution of Hydration Level")
plt.show()
```

```python
# Plot Water intake
plt.figure()
plt.hist(df["Daily Water Intake"].dropna())
plt.xlabel("Daily Water Intake")
plt.ylabel("Count")
plt.title("Distribution of Daily Water Intake")
plt.show()
```
